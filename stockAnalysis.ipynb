```python
import sqlalchemy as db
import datetime as dt
import pandas as pd
import talib
from tabulate import tabulate
import mplfinance as mpf


def getSqliteFilePath():
    return 'bseData.Sqlite'


def getStockData(script):
    sqliteFilePath = getSqliteFilePath()
    engine = db.create_engine('sqlite:///'+sqliteFilePath)
    connection = engine.connect()
    metadata = db.MetaData()
    table = db.Table(script, metadata, autoload=True, autoload_with=engine)
    query = db.select([table])
    result_proxy = connection.execute(query)
    result_set = result_proxy.fetchall()
    data = pd.DataFrame(result_set, columns=[
                        'Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
    return data


df = getStockData('CMSINFO.NS')
df['MA200'] = talib.SMA(df['Close'], timeperiod=200)
# print(df.columns)
#kws = dict(volume=True, tight_layout=True)
mpf.plot(df.set_index('Date').tail(100), style='yahoo', type='hollow_and_filled', volume=True, mav=(50, 20), figscale=2)
```
